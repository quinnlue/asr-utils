```python
import pandas as pd
from metric import score_wer
```

```python
df = pd.read_csv("output_combo.csv")

def get_wers(df):
    ref = df['ref'].tolist()
    pred = df['pred'].tolist()
    wers = []
    for i in range(len(ref)):
        r, p = ref[i], pred[i]
        if not isinstance(r, str) or not isinstance(p, str):
            wers.append(float('nan'))
        else:
            wers.append(score_wer(r, p))
    return wers

df['wers'] = get_wers(df)



```

```python
# df = pd.read_csv("output.csv")
df['weighted_wer'] = df['wers'] * df['len']
df = df.sort_values(by="weighted_wer", ascending=False)
# df = df.iloc[16:]
df['weighted_wer'].sum()/df['len'].sum()
```
